What is Heap Data Structure?
A heap is a binary tree-based data structure that satisfies the heap property: the value of each node is greater than or equal to the value of its children. This property makes sure that the root node contains the maximum or minimum value (depending on the type of heap), and the values decrease or increase as you move down the tree.

**here, heap property means - Property of a node in which key of every parent node need to be lesser than or equal to OR greater than or equal to the child node's key.**

There are two main types of heaps: min-heap and max-heap. 

In a min-heap, for every node i other than the root, the value of the node must be greater than or equal to the value of its parent node. In other words, the minimum value is at the root, and the values increase as you move down the tree.

In a max-heap, for every node i other than the root, the value of the node must be less than or equal to the value of its parent node. In other words, the maximum value is at the root, and the values decrease as you move down the tree.


here are examples of three min heaps and three max heaps:

**Min Heap Examples:**

1. Example 1:
   ```
            1
          /   \
         2     3
       / \   / \
      4   5 6   7
   ```
   In this min heap, every parent node has a value less than or equal to the values of its children.

2. Example 2:
   ```
            3
          /   \
         5     7
       / \   / \
      8   9 10 11
   ```
   Another example of a min heap where the parent nodes have values less than or equal to their children.

3. Example 3:
   ```
             5
           /   \
          8     10
        /  \   /  \
       11   15 14  20
   ```
   This min heap satisfies the min-heap property, where each parent node has a value less than or equal to its children.

**Max Heap Examples:**

1. Example 1:
   ```
            7
          /   \
         6     5
       / \   / \
      4   3 2   1
   ```
   In this max heap, every parent node has a value greater than or equal to the values of its children.

2. Example 2:
   ```
            10
          /   \
         9     8
       / \   / \
      7   6 5   4
   ```
   Another example of a max heap where the parent nodes have values greater than or equal to their children.

3. Example 3:
   ```
             20
           /   \
          15    10
        /  \   /  \
       14   8 5   2
   ```
   This max heap satisfies the max-heap property, where each parent node has a value greater than or equal to its children.

These examples illustrate the min-heap and max-heap properties, where the minimum or maximum value is always at the root of the heap, and the values follow the heap property throughout the heap.
